```python
import matplotlib.pyplot as plt
```

```python
#titles and lablels
x =[1,2,3]
y =[5,7,4]
plt.plot(x,y)
plt.xlabel('Plot number')
plt.ylabel(' y ')
plt.title('Interesting Graph \n :) \n ')
plt.show()
```

```python
#legends
x2 = [1,2,3]
y2 = [10,14,16]
plt.plot(x, y,label='First Line')
plt.plot(x2,y2,label='Second Line')
plt.ylabel(' y ')
plt.title('Interesting Graph \n :) \n ')
plt.legend()
plt.show()
```

```python
#Bar charts
x = [2,4,6,10]
y = [6,7,8,10]

x2 =[1,2,3,4,5,6]
y2 =[7,9,2,3,5,4]
plt.bar(x,y,label='Bars_1' , color='r')
plt.bar(x2,y2,label='Bars_2',color ='c')
plt.show()

```

```python
#histogram - distribution
population_ages = [22,35,62,21,22,35,42,4,42,102,99,110,120,130,11,112,80,75,44,43,48,49]
bins = [10,15,20,30,40,50,60,70,80,90,100,120,130]
plt.hist(population_ages , bins , histtype='bar' , rwidth=0.8)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Interesting Graph \n :) \n ')
plt.show()
```

```python
#scatter
x =[1,2,3,4,5,6,7,8,9]
y =[5,8,4,3,1,8,9,6,4]
#s -> size
plt.scatter(x,y,label='scatter',color='k' , marker='*', s=100)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Interesting Graph \n :) \n ')
plt.legend()
plt.show()
```
